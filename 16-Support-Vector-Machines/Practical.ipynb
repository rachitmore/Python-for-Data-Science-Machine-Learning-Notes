```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data.keys()
```

```python
data['DESCR']
```

```python
df = pd.DataFrame(data['data'],columns=data['feature_names'])
```

```python
df.head()
```

```python
df.info()
```

```python
data["target_names"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df
y = data['target']
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model= SVC().fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C':[0.1,0.3,1,3,10,30,100,300,1000,3000,4000,5000,10000,30000],
    'gamma':[0.00003,0.00001,0.000003,0.000001,0.0000003,0.0000001,0.00000003,0.00000001,0.000000003,0.000000001]
}
```

```python
grid = GridSearchCV(SVC(),param_grid=param_grid,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.best_estimator_
```

```python
predictions = grid.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
